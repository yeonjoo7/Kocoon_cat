```python
import pandas as pd
```

```python
# 서울시 구별 CCTV 현황 파일 불러오기
# 자료 출처: http://data.seoul.go.kr/dataList/OA-2722/S/1/datasetView.do "서울 열린데이터 광장"

cctv = pd.read_excel('서울시 자치구 목적별 CCTV 설치 현황.xlsx')
cctv.head(5)
```

```python
# 오류를 방지하기 위해 불러온 dataframe에 column명 영문으로 바꾸기

cctv.rename(columns = {'기관명' : 'region'}, inplace = True)
cctv.rename(columns = {'소계' : 'cctv_total'}, inplace = True)
cctv.rename(columns = {'방범용' : 'cctv_crime_prevent'}, inplace = True)
cctv.rename(columns = {'도시공원, 놀이터' : 'cctv_park'}, inplace = True)
cctv.rename(columns = {'어린이 보호' : 'cctv_kids'}, inplace = True)
cctv.rename(columns = {'쓰레기 무단투기' : 'cctv_life'}, inplace = True)
cctv.rename(columns = {'기타' : 'cctv_other'}, inplace = True)
cctv.head()
```

```python
# 필요한 서울시 자치구명, CCTV 갯수 합계, 어린이 보호용 CCTV 합계 column만 선택하여 새로운 dataframe 만들기

cctv_rev = cctv[["region", "cctv_total","cctv_kids"]]
cctv_rev.head()
```

```python
# 인구 밀도 및 보호구역 파일 불러오기
# 두 dataframe을 합본하기 위해서 column명을 같게 만들기

protect = pd.read_excel('서울_보호구역추가.xlsx')
protect = protect.rename(columns={"gu":"region"})
protect.head()
```

```python
# 서울시 구별 CCTV 현황과 인구 밀도 및 보호구역 dataframe을 합본

seoul_cctv = protect.merge(cctv_rev, how='left')
seoul_cctv.head()
```
