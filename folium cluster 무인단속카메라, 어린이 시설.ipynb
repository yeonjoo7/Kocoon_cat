```python
import folium
import pandas as pd
import json
```

```python
# 전국무인교통단속카메라표준데이터 파일에서 서울 데이터만 추출하기
# 출처 : https://www.data.go.kr/data/15028200/standard.do "공공데이터포털"

camera = pd.read_csv("전국무인교통단속카메라표준데이터.csv", encoding="euc=kr")
seoul_position = camera[camera["시도명"].str.contains("서울")]
```

```python
# 서울시 무인단속 카메라 dataframe 영문명 변수로 변환

speed_cam = seoul_position[["시군구명","위도","경도"]]
speed_cam = speed_cam.rename(columns={"시군구명":"region", "위도":"lat", "경도":"lon"})
speed_cam.head()
```

```python
# Null 값 제거

speed_cam = speed_cam[speed_cam['lon'].isnull() == False]
speed_cam = speed_cam[speed_cam['lat'].isnull() == False]
```

```python
# 서울시 중앙 좌표 map에 표현
# Circle이 잘 보일 수 있도록 cartodbpositron 배경 스타일 적용

map = folium.Map(location = [37.564346, 126.982062], zoom_start=11,
                 tiles='cartodbpositron')
```

```python
# MarkerCluster 모듈 import
# 무인단속카메라 좌표로 MarkerCluster표현

from folium.plugins import MarkerCluster

mc = MarkerCluster()
for row in speed_cam.itertuples():
    mc.add_child(folium.Marker(location=[row.lat, row.lon])) 
    map.add_child(mc) 
    
map
```


## 어린이 보호 구역 마커클러스터 추가

```python
# 어린이 보호 구역 자료 불러오기 및 서울 데이터만 추출
# 출처 : https://www.data.go.kr/tcs/dss/selectStdDataDetailView.do  "공공데이터포털"

kids_protect = pd.read_csv("전국어린이보호구역표준데이터.csv", encoding="euc=kr")
kids_prot = kids_protect[["위도","경도","제공기관명"]]
kids_prot = kids_prot[kids_prot['제공기관명'].str.contains("서울")]
```

```python
# dataframe의 변수명 영문 변환 및 Null 값 제거

kids_prot = kids_prot.rename(columns={"제공기관명":"region", "위도":"lat", "경도":"lon"})

kids_prot = kids_prot[kids_prot['lon'].isnull() == False]
kids_prot = kids_prot[kids_prot['lat'].isnull() == False]

kids_prot.head()
```

```python
# 서울시 중심부의 seoul_map을 만듭니다.

seoul_map = folium.Map(location = [37.564346, 126.982062], zoom_start=11,
                 tiles='cartodbpositron')
```

```python
# 어린이 보호 구역 좌표로 MarkerCluster표현

from folium.plugins import MarkerCluster

kidsProt = MarkerCluster()
for row in kids_prot.itertuples():
    kidsProt.add_child(folium.Marker(location=[row.lat, row.lon])) 
    seoul_map.add_child(kidsProt) 

seoul_map
```
